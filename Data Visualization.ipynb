```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.bar(x,y,0.5,color='red',label='2015')
plt.bar(x+0.2,y**1.5,color='green',label='2016')
plt.xlabel('Numbers')
plt.ylabel('squares')
plt.title('Testing')
plt.legend()
plt.show()
labels=['Maths','phy','chem','eng','c++']
plt.pie(y,labels=labels,explode=(1,0,0,0,0),shadow='1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mean=np.array([0.0,0.0])
cov=np.array([[1.0,0.0],[0.5,1]])

dist=np.random.multivariate_normal(mean,cov,500)
plt.scatter(dist[:,0],dist[:,1])
mean2=np.array([5.0,6.0])
cov2=np.array([[1.0,0.5],[0.0,0.5]])

dist2=np.random.multivariate_normal(mean2,cov2,500)
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
